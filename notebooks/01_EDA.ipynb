# Exploratory Data Analysis (EDA)

This notebook covers data summarization, quality assessment, univariate/bivariate analysis, geographic trends, and outlier detection.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/insurance_data.csv')
df.head()
```

## Data Summarization & Quality Assessment

```
print(df.info())
print(df.describe())
print("Missing values:\n", df.isnull().sum())
```

## Univariate & Bivariate Analysis

```
plt.figure(figsize=(10, 6))
sns.histplot(df['Premium'], kde=True)
plt.title('Distribution of Premium')
plt.show()
```

```
plt.figure(figsize=(10, 6))
sns.boxplot(x='Province', y='TotalClaims', data=df)
plt.title('Total Claims by Province')
plt.show()
```

## Loss Ratio Analysis

```
df['LossRatio'] = df['TotalClaims'] / df['Premium']
plt.figure(figsize=(10, 6))
sns.barplot(x='VehicleType', y='LossRatio', data=df)
plt.title('Loss Ratio by Vehicle Type')
plt.show()
```
