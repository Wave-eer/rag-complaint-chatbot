# Task 3: A/B Hypothesis Testing

In this notebook, we test statistical hypotheses to determine if there are significant risk differences across various demographics (Province, ZipCode, Gender).

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

df = pd.read_csv('../data/raw/insurance_data.csv')
df['ProfitMargin'] = df['TotalPremium'] - df['TotalClaims']
df.head()
```

### Hypothesis 1: Risk Differences Across Provinces
**Null Hypothesis (H0):** There are no significant differences in risk (ProfitMargin) across provinces.

**Alternative Hypothesis (H1):** At least one province has a significantly different risk profile.

```python
provinces = df['Province'].unique()
groups = [df[df['Province'] == p]['ProfitMargin'] for p in provinces]

# Using ANOVA (assuming normality for large N)
f_stat, p_val = stats.f_oneway(*groups)

print(f"ANOVA F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Reject Null Hypothesis: There are significant risk differences across provinces.")
else:
    print("Fail to reject Null Hypothesis: No significant risk differences across provinces.")
```

### Hypothesis 2: Risk Differences Across Zip Codes
**Null Hypothesis (H0):** There are no significant differences in risk (ProfitMargin) across Zip Codes.

```python
zipcodes = df['ZipCode'].unique()
groups_zip = [df[df['ZipCode'] == z]['ProfitMargin'] for z in zipcodes]

# Kruskal-Wallis Test (Non-parametric alternative)
h_stat, p_val_zip = stats.kruskal(*groups_zip)

print(f"Kruskal-Wallis H-statistic: {h_stat:.4f}")
print(f"P-value: {p_val_zip:.4f}")

if p_val_zip < 0.05:
    print("Reject Null Hypothesis: There are significant risk differences across Zip Codes.")
else:
    print("Fail to reject Null Hypothesis: No significant risk differences across Zip Codes.")
```

### Hypothesis 3: Risk Differences Across Gender
**Null Hypothesis (H0):** There are no significant differences in risk (ProfitMargin) between Male and Female clients.

```python
male_risk = df[df['Gender'] == 'Male']['ProfitMargin']
female_risk = df[df['Gender'] == 'Female']['ProfitMargin']

# Independent T-Test
t_stat, p_val_gender = stats.ttest_ind(male_risk, female_risk, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val_gender:.4f}")

if p_val_gender < 0.05:
    print("Reject Null Hypothesis: There is a significant risk difference between genders.")
else:
    print("Fail to reject Null Hypothesis: No significant risk difference between genders.")
```
