### Visual Analysis

```python
plt.figure(figsize=(10,6))
sns.histplot(df["TotalPremium"])
plt.title("Total Premium Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="TotalPremium", y="TotalClaims", hue="ZipCode")
plt.title("TotalPremium vs TotalClaims by ZipCode")
plt.show()
```

### Geographic Trends

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Province", y="TotalPremium")
plt.title("Premium by Province")
plt.show()
```

### Outlier Detection

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, y="TotalClaims")
plt.title("Outlier Detection for TotalClaims")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, y="CustomValueEstimate")
plt.title("Outlier Detection for CustomValueEstimate")
plt.show()
```
