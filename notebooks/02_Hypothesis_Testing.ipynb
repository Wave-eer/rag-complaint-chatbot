# A/B Hypothesis Testing

Testing 4 null hypotheses:
1. No risk differences across provinces
2. No risk differences between zip codes
3. No significant margin differences between zip codes
4. No significant risk difference between women and men

```
import pandas as pd
from scipy import stats

df = pd.read_csv('../data/insurance_data.csv')
```

## 1. Risk differences across provinces (ANOVA / Chi-squared)

```
# Example ANOVA for TotalClaims across Provinces
groups = [group['TotalClaims'].values for name, group in df.groupby('Province')]
f_stat, p_val = stats.f_oneway(*groups)
print(f"F-statistic: {f_stat:.4f}, p-value: {p_val:.4f}")
```
